```python
from __future__ import annotations

import sys
from pathlib import Path


def find_project_root(start: Path | None = None) -> Path:
    current = (start or Path.cwd()).resolve()
    for candidate in [current, *current.parents]:
        if (candidate / "src").exists() and (candidate / "data" / "raw" / "WA_Fn-UseC_-HR-Employee-Attrition.csv").exists():
            return candidate
    raise RuntimeError("Project root not found. Open the notebook from the repository or one of its subfolders.")


PROJECT_ROOT = find_project_root()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

PROJECT_ROOT
```

## Objective

Review the statistical workforce evidence exported by the companion R scripts. This notebook focuses on interpretation-ready tables: numeric group tests, categorical associations, attrition correlations, and segment-level comparisons.

```python
import pandas as pd

metrics_dir = PROJECT_ROOT / "outputs" / "metrics"

numeric_tests = pd.read_csv(metrics_dir / "r_numeric_group_tests.csv")
categorical_tests = pd.read_csv(metrics_dir / "r_categorical_association_tests.csv")
correlations = pd.read_csv(metrics_dir / "r_numeric_attrition_correlations.csv")
group_comparisons = pd.read_csv(metrics_dir / "r_all_attrition_group_comparisons.csv")

{
    "numeric_tests": numeric_tests.shape,
    "categorical_tests": categorical_tests.shape,
    "correlations": correlations.shape,
    "group_comparisons": group_comparisons.shape,
}
```

## Numeric Group Test Evidence

```python
(
    numeric_tests
    .assign(abs_effect=lambda frame: frame["cohens_d"].abs())
    .sort_values(["p_value", "abs_effect"], ascending=[True, False])
    .drop(columns="abs_effect")
    .head(10)
)
```

## Categorical Association Evidence

```python
categorical_tests.sort_values("p_value").head(10)
```

## Numeric Correlation Ranking

```python
correlations.assign(
    absolute_correlation=lambda frame: frame["correlation_with_attrition"].abs()
).sort_values("absolute_correlation", ascending=False)
```

## Highest Observed Attrition Segments

```python
(
    group_comparisons
    .query("employees >= 20")
    .sort_values(["attrition_rate", "employees"], ascending=[False, False])
    .head(20)
)
```
